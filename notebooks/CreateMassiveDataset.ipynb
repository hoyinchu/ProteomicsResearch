```python
%load_ext autoreload
%autoreload 2
import numpy as np
import pandas as pd
import os
from util import read_json_from, write_json_to, calc_eta
from data import proteomeHD_df
from visualization_util import df_roc_analysis,df_precision_recall_analysis
from validation import validate_relations_in_df,full_report_calculation_by_size,simplifed_report_df,is_corum_verifiable
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
PATH_ROOT = os.getcwd().replace("\\","/").replace("/notebooks","")
```

```python
proHD_coregulation_scores = pd.read_csv(f"{PATH_ROOT}/data_sources/ProteomeHD/coregulation_scores_simplified.csv")
```

```python
prohd_ratios_min95_path = f"{PATH_ROOT}/data_sources/ProteomeHD/prohd_ratios_min95.csv"
prohd_ratios_min95_df = pd.read_csv(prohd_ratios_min95_path)
prohd_ratios_min95_df_simplified_id = prohd_ratios_min95_df['Simplified_protein_ID'].to_numpy()
prohd_ratios_min95_df_feature_matrix = prohd_ratios_min95_df.iloc[:,2:].fillna(0).to_numpy()
prohd_ratios_min95_df_lookup = {prohd_ratios_min95_df_simplified_id[i]:i for i in range(len(prohd_ratios_min95_df_simplified_id))}
proHD_cols = [col for col in proteomeHD_df.columns if 'Ratio' in col]
```

```python
p1s = proHD_coregulation_scores['protein_1'].to_numpy()
p2s = proHD_coregulation_scores['protein_2'].to_numpy()
coregs = proHD_coregulation_scores['coregulation_score'].to_numpy()
```

```python
len(coregs)
```

```python
from util import append_to_csv
big_chungus_file_name = f"{PATH_ROOT}/data_sources/ProteomeHD/proteomeHD_all_pairs_combined.csv"
big_chungus_cols = ['combined_name','protein_1','protein_2','coregulation_score','binary_interaction','cocomplex']
big_chungus_cols = big_chungus_cols + proHD_cols
append_to_csv(big_chungus_file_name,big_chungus_cols,[])
```

```python
import time
from util import calc_eta, append_to_csv
from validation import eval_relation
chungus = []
starttime=time.time()
for i in range(len(p1s)):
    p1 = p1s[i]
    p2 = p2s[i]
    coreg = coregs[i]
    row_to_append = [p1+'-'+p2,p1,p2,coreg]
    relation_report = eval_relation(p1,p2)
    if relation_report['experiments'] > 0 or relation_report['database'] > 0:
        row_to_append.append(1)
    else:
        row_to_append.append(0)
    if relation_report['cocomplex'] > 0:
        row_to_append.append(1)
    else:
        row_to_append.append(0)
    vec_1 = prohd_ratios_min95_df_feature_matrix[prohd_ratios_min95_df_lookup[p1]]
    vec_2 = prohd_ratios_min95_df_feature_matrix[prohd_ratios_min95_df_lookup[p2]]
    comb_vec = vec_1+vec_2
    row_to_append = row_to_append + comb_vec.tolist()
    chungus.append(row_to_append)
    if (i!=0 and i % 100000 == 0) or i == len(p1s)-1:
        calc_eta(starttime,i,len(p1s))
        append_to_csv(big_chungus_file_name,None,chungus)
        chungus.clear()
```
